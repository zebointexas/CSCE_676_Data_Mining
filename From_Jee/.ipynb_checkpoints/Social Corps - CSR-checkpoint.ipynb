```python
# your code here 
# clean data 
# convert cleaned durations to seconds

# ignore >, < signs in front, or space between numbers and units
data = pd.read_csv('output.csv')
cleanData = pd.DataFrame()
for index, row in data.iterrows():
    # All regex below ignore cases with (?i) 
    
    # symbol (>, <, ~) + digits (+ space(s)) + time unit
    space_pattern = '(?i)[^\d\w]*(\d+)\s*(s|m|h|d|w)\w*?'
    
    # range of numbers : digits (- , to, or) digits + time unit
    avg_pattern = '(?i)[^\d\w]*(\d+)\s*(\-|to|or)\s*(\d+)\s*(s|m|h|d|w)\w*'
    
    # symbol + digits + symbol + time unit
    rest_pattern = '(?i)[^\d]+(\d+)[^\d\w]*\s*(s|m|h|d|w)\w*'
    
    # digits + symbol + time unit : ex) 1.5 minutes rounded to 1 minute
    num_pattern_1 = '(?i)(\d+)[^\d\w]*\s*(s|m|h|d|w)\w*'
    
    # detect only numbers : default as min
    num_pattern_2 = '(?i)[^\d\w]*(\d+)[^\d\w]*'
    
    # numbers with : between.   50:25 , 1:10    -> default as min:sec
    third_pattern = '(?i)(\d+)\s*\:\s*(\d+)\s*\w*'
    
    # detects numbers after :     -> default as sec
    fourth_pattern = '(?i)\d*\s*\:\s*(\d+)\s*\w*'
    
    # a few minute = 1 min ,,    approximately one minute = 1 min
    char_pattern_1 = '(?i)(\w*)\s*(\w*)\s*(s|m|h|d|w)\w*'
    
    detected = False
    if re.match(space_pattern, str(row[4])):
        matches = re.findall(space_pattern, str(row[4]))
        num = int(matches[0][0])
        unit = matches[0][1]
        detected = True
    elif re.match(avg_pattern, str(row[4])):
        matches = re.findall(avg_pattern, str(row[4]))
        start = int(matches[0][0])
        end = int(matches[0][2])
        unit = matches[0][3]
        num = (start + end) / 2
        detected = True
    elif re.match(rest_pattern, str(row[4])):
        matches = re.findall(rest_pattern, str(row[4]))
        num = int(matches[0][0])
        unit = matches[0][1]
        detected = True
    elif re.match(num_pattern_1, str(row[4])):
        matches = re.findall(num_pattern_1, str(row[4]))
        num = int(matches[0][0])
        unit = matches[0][1]
        detected = True
    elif re.match(num_pattern_2, str(row[4])):
        matches = re.findall(num_pattern_2, str(row[4]))
        num = int(matches[0][0])
        unit = 'm'
        detected = True
    elif re.match(third_pattern, str(row[4])):
        matches = re.findall(third_pattern, str(row[4]))
        num = int(matches[0][1])*60 + int(matches[0][0])
        unit = 's'
        detected = True
    elif re.match(fourth_pattern, str(row[4])):
        matches = re.findall(fourth_pattern, str(row[4]))
        num = int(matches[0][0])
        unit = 's'
        detected = True
    elif re.match(char_pattern_1, str(row[4])):
        matches = re.findall(char_pattern_1, str(row[4]))
        num = 1
        unit = matches[0][2]
        detected = True
        
    if detected:
        if unit == 's' or unit == 'S':
            row[4] = int(num)
        elif unit == 'm' or unit == 'M':
            row[4] = int(num * 60)
        elif unit == 'h' or unit == 'H':
            row[4] = int(num * 3600)
        elif unit == 'd' or unit == 'D':
            row[4] = int(num * 3600 * 24)
        else:
            row[4] = int(num * 3600 * 24 * 7)
        cleanData = cleanData.append(row)
```

```python
# your code here, including all the helpful print statements
# when you run this block, we want to see all of your intermediate steps
# you can save the rules you discover for printing in the following cells (this will help us grade by
# keeping these separate)
import pandas as pd
import math

movie_rating_df = pd.read_csv('./movie_rated.csv')
movie_rating_df = movie_rating_df.sort_values(by=['userId','movieId'], ascending=[True, True])

# Finding 1-itemset
item_sets = dict()
baskets = 0
for index, row in movie_rating_df.iterrows():
    movie_id = row[1]
    count = item_sets.get(movie_id, 0) + 1
    item_sets[movie_id] = count
    if row[0] > baskets:
        baskets = row[0]

minsup = math.ceil(baskets * 0.2)
freq_item_sets = []
first_itemset_count = []
for item in item_sets:
    if item_sets[item] >= minsup:
        freq_item_sets.append(item)
        first_itemset_count.append(item_sets[item])

freq_item_sets.sort()
print('Frequent 1-Itemsets:')
print(freq_item_sets)
print()

# Finding 2-itemsets
pairs = dict()
for i in range(len(freq_item_sets)):
    i_dict = dict()
    for j in range(i+1, len(freq_item_sets)):
        i_dict[freq_item_sets[j]] = 0
    pairs[freq_item_sets[i]] = i_dict

freq_set = set(freq_item_sets)
user = 1
movie_list = []
for index, row in movie_rating_df.iterrows():
    if user != row[0]:
        movie_list = []
        user = row[0]
    movie = row[1]
    if movie in freq_set:
        for item in movie_list:
            pairs[item][movie] += 1
        movie_list.append(movie)

f_k = []
counts = []
for first_movie in pairs:
    for second_movie in pairs[first_movie]:
        if pairs[first_movie][second_movie] >= minsup:
            f_k.append([first_movie, second_movie])
            counts.append(pairs[first_movie][second_movie])
print('Frequent 2-Itemsets:')
print(f_k)
print()

# generates c_(k+1) given f_k
def generate_c(f_k):
    c_k = []
    for i in range(len(f_k)):
        for j in range(i+1, len(f_k)):
            merge = True
            for k in range(len(f_k[i])-1):
                if f_k[i][k] != f_k[j][k]:
                    merge = False
                    break
            if merge:            
                new_cand = f_k[i].copy()
                new_cand.append(f_k[j][-1])
                c_k.append(new_cand)
    return c_k

# hash tree with degree 3
r = int(len(freq_item_sets)/3)
left = set()
bottom = set()
right = set()

# hash function
for i in range(r):
    left.add(freq_item_sets[i])
for i in range(r, 2*r):
    bottom.add(freq_item_sets[i])
for i in range(2*r, len(freq_item_sets)):
    right.add(freq_item_sets[i])

# node class for tree 
class Node:
    def __init__(self, data, count, left=None, bottom=None, right=None):
        self.data = data
        self.count = count
        self.left = left
        self.bottom = bottom
        self.right = right

# add all itemsets in c_k to the tree's leaves
# a leaf can have multiple candidates in node.data
# its counts are stored in node.count (same index)
# returns root and nodes
# nodes contains reference to leaves
def make_tree(c_k):
    root = Node([],[])
    nodes = []
    for cand in c_k:
        curr = root
        for i in range(len(cand)):
            movie = cand[i]
            if movie in left:
                if curr.left == None:
                    curr.left = Node([],[])
                curr = curr.left
            elif movie in bottom:
                if curr.bottom == None:
                    curr.bottom = Node([],[])
                curr = curr.bottom
            elif movie in right:
                if curr.right == None:
                    curr.right = Node([],[])
                curr = curr.right
        if len(curr.data) == 0:
            nodes.append(curr)
        curr.data.append(cand)
        curr.count.append(0)
    return root, nodes

# helper function I used to debug
def print_tree(root):
    if root != None:
        print(root.data)
        print_tree(root.left)
        print_tree(root.bottom)
        print_tree(root.right)

# find index from f_k (='count' in this function) given data
def find_index(count, data):
    for index in range(len(count)):
        lst = count[index]
        if len(data) == len(lst):
            found = True            
            for i in range(len(lst)):
                if data[i] != lst[i]:
                    found = False
                    break;
            if found:
                return index
    return -1

# recursively find the leaves that transaction can end up to.
# add 1 to all counts that this function reaches at
def add_count(transaction, root, start, data, k):
    if k == 0:
        index = find_index(root.data, data)
        if index >= 0:
            root.count[index] = root.count[index] + 1
    else: 
        for i in range(start, len(transaction)):
            if transaction[i] in left and root.left != None:
                new_data = data.copy()
                new_data.append(transaction[i])
                add_count(transaction, root.left, i+1, new_data, k-1)
            elif transaction[i] in bottom and root.bottom != None:
                new_data = data.copy()
                new_data.append(transaction[i])
                add_count(transaction, root.bottom, i+1, new_data, k-1)
            elif transaction[i] in right and root.right != None:
                new_data = data.copy()
                new_data.append(transaction[i])
                add_count(transaction, root.right, i+1, new_data, k-1)

k = 2
f_all = []
for itemset in freq_item_sets:
    f_all.append([itemset])
counts_all = first_itemset_count.copy()

# while loop until max_itemset is found
while len(f_k) > 0:
    f_all.extend(f_k)
    counts_all.extend(counts)
    user = movie_rating_df.iloc[0][0]
    k += 1
    transaction = []
    c_k = generate_c(f_k)
    root, nodes = make_tree(c_k)
    for index, row in movie_rating_df.iterrows():
        if user != row[0]:
            user = row[0]
            add_count(transaction, root, 0, [], k)
            transaction = []
        transaction.append(row[1])
    add_count(transaction, root, 0, [], k)
    f_k = []
    counts = []
    for node in nodes:
        for i in range(len(node.count)):
            if node.count[i] >= minsup:
                f_k.append(node.data[i])
                counts.append(node.count[i])
    print("Frequent " + str(k) + "-Itemsets:")
    print(f_k)
    print()

# returns permutation of subsets + their complement subsets for given itemset
def get_permuts(itemset):
    permuts = []
    compls = []
    for i in range(len(itemset)):
        compl = [itemset[i]]
        permut = itemset.copy()
        del permut[i]
        permuts.append(permut)
        compls.append(compl)
        permuts.append(compl)
        compls.append(permut)
    return permuts, compls

# save association rules in rules dict()
rules = dict()
for i in reversed(range(len(f_all))):
    itemset = f_all[i]
    sup_total = counts_all[i]
    if len(itemset) <= 1:
        break
    permuts, compls = get_permuts(itemset)
    for i in range(len(permuts)):
        conf = sup_total / counts_all[find_index(f_all, permuts[i])]
        if conf >= 0.8:
            rules[str(permuts[i])] = str(compls[i])

print(rules)

    


    
```
