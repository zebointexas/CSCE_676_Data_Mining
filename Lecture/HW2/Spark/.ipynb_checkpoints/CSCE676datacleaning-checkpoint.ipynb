## Presidential Cash Money

We're going to do a quick dive into contributions to presidential candidates, as maintained by the Federal Elections Commission: http://www.fec.gov/disclosurep/PDownload.do

Specifically, we'll focus on contributions from Texas in the 2016 cycle: ftp://ftp.fec.gov/FEC/Presidential_Map/2016/P00000001/P00000001-TX.zip

Some code from Python for Data Analysis by McKinney.

### Let's get this started

```python
# let's start by reading in the latest Federal Elections Commission (FEC)
# list of contributions to presidential candidates. For this analysis, we'll
# focus solely on contributions made by Texans.
import pandas as pd
data_path='./data/2020-P00000001-TX.csv'
contributions = pd.read_csv(data_path, index_col=False, low_memory=False)

# .head() is a helper function for displaying the top-most rows of a pandas dataframe
contributions.head()
```

```python
# since there are many columns of little interest to us, let's
# go ahead and drop them altogether
contributions.drop(['cmte_id', 'cand_id', 'form_tp', 'file_num', 'tran_id', 'election_tp', 'receipt_desc', 'memo_cd', 'memo_text'], axis=1, inplace=True)

# also, since there are some negative contributions for some 
# bookkeeping (out of scope of our tutorial), we'll drop those
contributions = contributions[contributions.contb_receipt_amt > 0]
```

### Can we identify patterns in contributions?

Let's represent each occupation by a vector of the candidates that they have 
contributed to. Each element in the vector represents a particular candidate. Our hypothesis is that there might be patterns of contributions that connect **similar** occupations. Let's take a look:

```python
# first let's consider just contributions from the top-200 occupations
top_200_occupations = contributions.contbr_occupation.value_counts()[:200]
contributions_from_top_200_occupations = contributions[contributions.contbr_occupation.isin(top_200_occupations.index)].reset_index()
contributions_from_top_200_occupations.shape
```

Now we can look at the dollars contributed per occupation per candidate:

```python
# dollar sum for each occupation for each candidate
candidate_occupation_dollars = contributions_from_top_200_occupations.groupby(['contbr_occupation','cand_nm'])['contb_receipt_amt'].sum()

# dollar sum for each occupation
sum_occupation_dollars = contributions_from_top_200_occupations.groupby(['contbr_occupation'])['contb_receipt_amt'].sum()

# represent as ratio
ratio_per_occupation = candidate_occupation_dollars / sum_occupation_dollars
```

```python
candidate_occupation_dollars.head()
```

```python
sum_occupation_dollars
```

```python
ratio_per_occupation
```

```python
# convert the multi-level indexed series to dataframe
occupation_vectors = ratio_per_occupation.unstack(level=1)

# fill the NaN with 0
occupation_vectors.fillna(0,inplace=True)

occupation_vectors.head()
```

Perhaps some candidates correlate with each other from this **occupation** donor perspective. Let's take a look:

```python
occupation_vectors.corr()
```

Since there are so many candidates, we can project them into a lower-dimensional space using PCA, perhaps accounting for this observed correlation.

```python
from sklearn.decomposition import PCA
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD()
# dimension reduction (originally 24 dimensions / candidates)
pca = PCA(n_components=4)
pca.fit(occupation_vectors)
reduced_occupation_vectors = pca.transform(occupation_vectors)
reduced_occupation_vectors
```

```python
import numpy as np
b=[item/sum(pca.explained_variance_ratio_) for item in pca.explained_variance_ratio_]
print (np.cumsum(b))
```

Finally, we can look for natural groups of occupations in this lower-dimensional space. What do we observe?

```python
from sklearn.cluster import KMeans
from collections import defaultdict

# cluster the occupations
c = KMeans(n_clusters=8, init='k-means++')
c.fit(reduced_occupation_vectors, y=None)

clusters = defaultdict(list)
for i in range(len(occupation_vectors.index)):
	clusters[str(c.labels_[i])].append(occupation_vectors.index[i])
```
