# Import data and Python Library

```python
import pandas as pd 
data = pd.read_csv("1.csv") 
data.head(1)  
```

```python
import numpy as np
from sklearn.preprocessing import normalize

####################################
x = data['Median']
x_min = min(data['Median'])
x_max = max(data['Median'])
x_mean = np.mean(data['Median'])
 
x_norm_Median = (x - x_min) / (x_max - x_min)
####################################
x = data['LWJ_rate']
x_min = min(data['LWJ_rate'])
x_max = max(data['LWJ_rate'])
x_mean = np.mean(data['LWJ_rate'])
 
x_norm_LWJ_rate = (x - x_min) / (x_max - x_min)
####################################
x = data['U_rate']
x_min = min(data['U_rate'])
x_max = max(data['U_rate'])
x_mean = np.mean(data['U_rate'])
 
x_norm_U_rate = (x - x_min) / (x_max - x_min)
####################################
x = data['FTY Ratio']
x_min = min(data['FTY Ratio'])
x_max = max(data['FTY Ratio'])
x_mean = np.mean(data['FTY Ratio'])
 
x_norm_FTY_Ratio = (x - x_min) / (x_max - x_min)
####################################
x = data['Part Time Ratio']
x_min = min(data['Part Time Ratio'])
x_max = max(data['Part Time Ratio'])
x_mean = np.mean(data['Part Time Ratio'])
 
x_norm_Part_Time_Ratio = (x - x_min) / (x_max - x_min)
####################################
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn import preprocessing


plt.rcParams.update({'font.size': 22})

fig, ax = plt.subplots(figsize=(20,18))

# LWJ_rate    U_rate    FTY Ratio     Part_Time_Ratio   
  
y = data['Major'][::-1]

x = x_norm_Median[::-1]
ax.plot(x,y,color="Red", linewidth=2, markersize=12, fillstyle="full")

x = x_norm_LWJ_rate  
ax.scatter(x,y,marker="4",color="Purple",s=1000)

x = x_norm_Part_Time_Ratio
ax.scatter(x,y,marker=">",color="Green",s=1000)

plt.title("Art majors income analysis")
 
plt.legend(('Income:27k - 50k', 'Low Wage Rate: 11%-26%', 'Part-Time Rate: 22%-46%'), loc='upper left', prop={'size': 22})

plt.show()
```
